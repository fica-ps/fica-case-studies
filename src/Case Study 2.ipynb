# Case Study with mixed signals
## This case study attempts to separate two mixed signals into their original form

```julia
#using Pkg
#Pkg.add("Plots")

using Plots

s1 = sin.((1:1000)./20)
plot(s1, linewidth=2, title = "S1")
```

```julia
s2 = vec(repeat(((1:200).-100)/100, 5))
plot(s2,linewidth=2,title="S2")
```

## Mixing the previous functions twice using different factors so we have enough data to apply the algorithm

```julia
#mixing matrix
A = [0.291 -0.5439 ; 0.6557 0.5572]
S = hcat(s1,s2)
X = S * A
```

```julia
plot(X[:,1], linewidth=2, title = "X1")
```

```julia
plot(X[:,2], linewidth=2, title = "X2")
```

## Data whitening, application of the FICA algorithm

```julia
include("fastICA.jl")
W = fastICA.whiten(X)
X1 =  X * W'
#fixed weights
iniwei = [1.02081 0.408655 ; -1.92523 -0.756068]

#direct call to deflation function to pass weights
a = fastICA.fast_ica_def(200,2,Array(X1'),0.0001,iniwei, 1.)
#a = fastICA.fast_ica(100,2,Array(X1'),0.0001,1.)
K = a * W
S = K * X'
```

## Plotting of results

```julia
plot(S[1,:,],linewidth=2, title = "S'1")
```

```julia
plot(S[2,:,],linewidth=2, title = "S'2")
```
